# Visualisation 

# Analysis by distribution of sales volume 

```python
# Histogram of Sales Volume
plt.figure(figsize=(10, 6))
plt.hist(df['Sales_Volume'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sales Volume')
plt.xlabel('Sales Volume')
plt.ylabel('Frequency')
plt.show()
```

# Analysis by Count of Cars in Region 

```python
# Countplot of Region
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Region')
plt.title('Count of Cars by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

# Analysis on Average price by Fuel type 

```python
# Bar plot of average Price_USD by Fuel_Type
plt.figure(figsize=(10, 6))
df.groupby('Fuel_Type')['Price_USD'].mean().plot(kind='bar', color='lightgreen')
plt.title('Average Price (USD) by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (USD)')
plt.show()
```

# Analysis on top models by Sales volume 

```python
# Group by Model and sum Sales_Volume, then sort and plot top 10
top_models = df.groupby('Model')['Sales_Volume'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_models.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Models by Total Sales Volume')
plt.xlabel('Model')
plt.ylabel('Total Sales Volume')
plt.xticks(rotation=45)
plt.show()
```

# Analysis on Sales volume by Region 

```python
# Group by Region and sum Sales_Volume
sales_by_region = df.groupby('Region')['Sales_Volume'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sales_by_region.plot(kind='bar', color='coral')
plt.title('Total Sales Volume by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales Volume')
plt.show()
```

# Analysis by Distribution of Sales volume 

```python
# Histogram of Sales Volume
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales_Volume'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sales Volume')
plt.xlabel('Sales Volume')
plt.ylabel('Frequency')
plt.show()
```

# Analysis by Sales volume Over time 

```python
# Group by Year and sum Sales_Volume
sales_over_time = df.groupby('Year')['Sales_Volume'].sum()

plt.figure(figsize=(10, 6))
sales_over_time.plot(kind='line', marker='o', color='blue')
plt.title('Total Sales Volume Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales Volume')
plt.show()
```

# Analysis on Average Price_USD by Model

```python
# Group by Model and calculate average Price_USD, then sort and plot top 10
avg_price_by_model = df.groupby('Model')['Price_USD'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_price_by_model.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Models by Average Price (USD)')
plt.xlabel('Model')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

# Analysis on Average Price_USD by Region 

```python
# Group by Region and calculate average Price_USD
avg_price_by_region = df.groupby('Region')['Price_USD'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_price_by_region.plot(kind='bar', color='coral')
plt.title('Average Price (USD) by Region')
plt.xlabel('Region')
plt.ylabel('Average Price (USD)')
plt.show()
```

# Analysis on Average Price_USD by Fuel_type 

```python
# Group by Fuel_Type and calculate average Price_USD
avg_price_by_fuel_type = df.groupby('Fuel_Type')['Price_USD'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_price_by_fuel_type.plot(kind='bar', color='blue')
plt.title('Average Price (USD) by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (USD)')
plt.show()
```

# Analysis by Price_USD Over time

```python
# Group by Year and calculate average Price_USD
avg_price_over_time = df.groupby('Year')['Price_USD'].mean()

plt.figure(figsize=(10, 6))
avg_price_over_time.plot(kind='line', marker='o', color='purple')
plt.title('Average Price (USD) Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.show()
```
